```python
menu = {"포도주스" : 800, "오렌지주스" : 800, "콜라" : 1000, '사이다': 1000, '몬스터' : 2000}


now_money = 0 #입력된 돈 초기 값 0
vending_number =0 #뽑은 음료수 개수



def order_drink(drink, n, money):
    """
    주문한 음료와 가격 계산하기
    매개변수 
        drink : 주문할 음료의 번호
        n : 주문할 음료의 갯수
        money : 
        
    """
    cost = 0
    global vending_money
    global vending_number
    global now_money
    
    d_name = list(menu.keys())[drink-1]
    d_cost = menu.get(d_name)
    
    cost = d_cost * n
    
    if cost > money:
        print("금액이 부족합니다.. 처음으로 돌아갑니다.")
        print(f"현재 금액 : {money}원")
        now_money = money
    else :
        return_money = money - cost
        print(f'\n[주문내역] 음료 : {d_name} 개수 : {n} , 가격 : {cost}원')
        print(f"자판기에 남은 돈 : {return_money}원")
        print("~~~~~~~~~[BYE]~~~~~~~~~~~~~")
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        
        now_money = return_money
        vending_number += n
        
        
        
def vendingmachine():
    '''
    주문할 음료와 갯수를 입력
    '''
    menu_order = ['포도주스', '오렌지주스', '콜라', '사이다','몬스터']
    menu_cost = [800,800,1000,1000,2000]
    global now_money
    
   
    print("\n======자판기메뉴=======")
    for i,(k,v) in enumerate(zip(menu_order,menu_cost), start=1):
        print(f'{i}) {k} \t \\{v}원' )
    
    print(f"현재 돈: {now_money}원")
    
    while True:
        drink = int(input("주문할 음료수를 입력해주세요(1~5):"))
        if 0 < drink < 6:
            try:
                number = int(input("개수를 입력해주세요 : "))
            except:
                print("개수를 다시 입력해주세요")
                continue
            else :
                order_drink(drink, number,now_money)
                break    
        else:
            print("올바른 음료 숫자를 입력해주세요.")
            continue


def admin_mode():
    '''
    관리자모드
    1. 추가
    2. 수정
    3. 삭제
    0. 처음입력으로 돌아가기
    
    미완성..
    '''
    def admin_add():
        print("[관리자 모드 추가]")
        pass
    
    def admin_mod():
        print("[관리자 모드 수정]")
        pass
    
    def admin_del():
        print("[관리자 모드 삭제]")
        pass
    
    while True:
        try:
            mode = int(input("\n[관리자 권한]\n실행할 모드를 입력해주세요. (0~3)"))
        except:
            print("숫자를 입력해주세요.")
            continue
        else:
            if 0 < mode < 4:
                if mode == 1:
                    admin_add()
                    continue
                elif mode == 2:
                    admin_mod()
                    continue
                elif mode == 3:
                    admin_del()
                    continue
            elif mode == 0:
                print("[관리자모드 종료...]")
                break
            else:
                print("0,1,2,3 중 하나 입력하세요.")
                continue
       
            
def main():  
    '''
    메인함수
    '''
    global now_money # 전역변수로 지정된 now_money의 값을 사용
    global menu
    
    while True:
        print("\n[자판기]")
        mode = input("버튼을 입력해주세요(0:종료, 1:자판기 이용, 2:현황) : ")
        if mode == "1":
            #자판기 이용
            print("*************자판기 이용하기***********")
            for key ,value in menu.items() :
                print(f'{key} \t \\{value}원')
            print("***************************************")

            if now_money != 0: # 기존에 돈이 남아있다면 출력
                print(f"!!!!!!!!!자판기에 남아있던 돈 : {now_money}원")

            try:
                menu_order = int(input("(1) 돈넣기 , (2) 음료수 뽑기 : "))
            except Exception as e:
                print("1,2 다시 입력하세요.")
                continue
            else:
                if menu_order == 1:
                    #돈넣기
                    while True:
                        try:
                            input_money = int(input("돈 넣기 : "))
                            print(f"투입된 돈: {input_money}원")
                            check_money = input("넣은 금액이 맞습니까? (y/n):")
                            if check_money =="y":
                                now_money += input_money
                                vendingmachine()
                                break   # 주문하고 끝나면 초기로 돌아가기
                            elif check_money == "n":
                                continue
                            else:
                                print("다시 입력해주세요")
                                continue
                        except :
                            print("돈(숫자)을 입력하세요")
                            continue
                elif menu_order == 2 :
                    #음료수 뽑기
                    if now_money == 0:
                        print("음료수를 뽑을 수 없습니다. 돈을 넣어주세요. 처음으로 돌아갑니다")
                        continue
                    else:
                        vendingmachine()
                else:
                    print("1,2 중 하나 다시 입력하세요.")
                    continue
        elif mode == "2" :
            #현황
            print("\n[주문이력]")
            print(f"뽑은 음료수 개수 : {vending_number}개")
            print(f"자판기에 남은 금액 : {now_money}원")
            continue
        elif mode == "0" :
            #종료
            break
        elif mode == "9999" :
            #관리자모드
            print("**********************관리자모드**********************")
            admin_mode() # 관리자 모드 함수 
        else:
            print("다시 입력하세요.")
        
        
if __name__=="__main__":
    main()
```
